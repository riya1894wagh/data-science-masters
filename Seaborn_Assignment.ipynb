### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans:


**Distplots**

Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.




```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5],hist=False)

plt.show()
```

**Line plot**

The line plot is one of the most basic plot in seaborn library.  This plot is mainly used to visualize the data in form of some time series, i.e. in continuous manner.

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.lineplot(x="total_bill",
             y="tip",
             hue="sex",
             data=tips)
```

**Lmplot**

The Lmplot is another one of the basic plots in the seaborn library. The Lmplot shows a line that represents a linear regression model with the data points on the given two-dimensional (2-D) space. In this 2-D space, we can set x and y variables as the vertical and horizontal labels, respectively.

```python
sns.lmplot(x="total_bill",
             y="tip",
             hue="sex",
             data=tips)
```

**Countplot**

The countplot is used to represent the occurrence(counts) of the observation present in the categorical variable.

It uses the concept of a bar chart for the visual depiction.

```python
sns.countplot(x='sex',data=tips)  
```

```python
sns.countplot(x='sex',hue='smoker',data=tips)  
```

**KDE Plot**

A Kernel Density Estimate (KDE) Plot is used to plot the distribution of continuous data.

 

```python
sns.kdeplot(x = 'total_bill' , data = tips )
```

**Scatter plot**

```python
sns.scatterplot(x='total_bill', y ='tip' ,
data = tips , hue = 'sex')
```

**Joint plot**

A Joint Plot is also used to plot the correlation between data.

```python
sns.jointplot(x='total_bill' , y ='tip',
data = tips , kind = 'reg')
```

```python
sns.jointplot(x='total_bill' , y ='tip',
data = tips , kind = 'scatter')
```

**kind** 

The kind of plot to be plotted. It can be one of the following.
'scatter', 'hist', 'hex', 'kde', 'reg', 'resid'

**Pair plots**

Seaborn lets us plot multiple scatter plots. It’s a good option when you want to get a quick overview of your data.

```python
sns.pairplot(tips)
```

**Heatmaps**

A heat map can be used to visualize confusion, matrices, and correlation.

```python
sns.heatmap(tips.corr())
```
